```python
import requests
from bs4 import BeautifulSoup
import re
import json

def crawl_page_content():
    url = 'https://namu.wiki/w/%EC%9C%A0%EB%82%98(%EC%86%8C%EB%93%9C%20%EC%95%84%ED%8A%B8%20%EC%98%A8%EB%9D%BC%EC%9D%B8)?from=%EC%8B%9C%EA%B2%8C%EB%AC%B4%EB%9D%BC%20%EC%9C%A0%EC%9A%B0%EB%82%98'

    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    # div data-v-0049117a 부분의 내용만 추출
    content = soup.find_all('div', {'data-v-0049117a': True})

    # 정규표현식 패턴 설정
    pattern = r"번 문단 \(이전 역사\) 문서의 r 판, "

    # 추출한 내용 출력
    crawled_texts = []
    for item in content:
        # 패턴에 맞는 문자열을 빈 문자열로 치환
        clean_text = re.sub(pattern, '', item.text)
        crawled_texts.append(clean_text)

    # 크롤링한 텍스트를 JSON 파일로 저장
    with open('crawled_text.jsonl', 'w', encoding='utf-8') as f:
        json.dump(crawled_texts, f, ensure_ascii=False, indent=4)

crawl_page_content()
```
